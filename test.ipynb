```python
import face_recognition
import numpy as np
from PIL import Image, ImageDraw
from IPython.display import display
```

```python
path = "known_faces"
```

```python
from PIL import Image, ImageDraw
from IPython.display import display

pil_im = Image.open(path+'/este_1.png')
display(pil_im)
```

```python
image_path = "known_faces/este_1.png"

# Open the image using Pillow
pil_image = Image.open(image_path)

# Print initial properties
print(f"Initial Image mode: {pil_image.mode}")
```

```python
if pil_image.mode != 'RGB':
    pil_image = pil_image.convert('RGB')
    print("Converted image to RGB mode")
```

```python
numpy_image = np.array(pil_image)
```

```python
if numpy_image.dtype != np.uint8:
    raise ValueError("Image is not 8-bit. Please use an 8-bit image.")
if len(numpy_image.shape) != 3 or numpy_image.shape[2] != 3:
    raise ValueError("Image is not in RGB format.")
```

```python
try:
    este_face_encoding = face_recognition.face_encodings(numpy_image)[0]
    print("Face encoding successful:", este_face_encoding)
except RuntimeError as e:
    print(f"RuntimeError: {e}")
```

```python
import cv2
```

```python
image_path = "known_faces/mama.jpeg"
opencv_image = cv2.imread(image_path)
```

```python
if opencv_image is None:
    raise ValueError("Image not loaded. Please check the file path.")
```

```python
rgb_image = cv2.cvtColor(opencv_image, cv2.COLOR_BGR2RGB)
```

```python
try:
    este_face_encoding = face_recognition.face_encodings(rgb_image)[0]
    print("Face encoding successful:", este_face_encoding)
except RuntimeError as e:
    print(f"RuntimeError: {e}")
```
